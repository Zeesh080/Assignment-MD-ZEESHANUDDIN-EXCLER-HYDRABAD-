```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("D:/Assignments/Basic stats - 1/Basic stats - 1/sales_data_with_discounts.csv")
Numerical_columns = df.select_dtypes(include='number').columns
Categorical_columns = df.select_dtypes(include=['category','object']).columns
```

```python
#Descriptive Analytics for Numerical Columns
for column in Numerical_columns:
    mean = df[column].mean()
    median = df[column].median()
    std = df[column].std()
    min = df[column].min()
    max = df[column].max()
    skw = df[column].skew()
    print(f'''{column} ->         Mean : {mean}
                      Median : {median} 
                      Std Dev : {std}
                      Minimun ; {min} 
                      Maximum : {max} 
                      Skewness : {skw}''')
    
```

```python
#	Histograms
for column in Numerical_columns:
    skewness = df[column].skew()
    if skewness > 0:
        print("The distribution is Right-skwed.")
    elif skewness < 0:
        print("The distribution is Left-skwed.")
    else:
        print("The distribution is Approximately symmetric.")
    print(f'Skewness for {column}: {skewness}')
    plt.hist(df[column],bins=30,edgecolor='k')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#	Boxplots
for column in Numerical_columns:
    sales_series = df[column]
    Q1 = sales_series.quantile(0.25)
    Q3 = sales_series.quantile(0.75)
    IQR = Q3-Q1
    print(f'Q1 : {Q1}')
    print(f'Q3 : {Q3}')
    print(f'Interquartile range : {IQR}')
    l_b=Q1-1.5*IQR
    u_b=Q3+1.5*IQR
    ext_values=sales_series[(sales_series<l_b)|(sales_series>u_b)]
    print("Extreme values found :")
    print(list(ext_values))
    plt.boxplot(df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#	Bar Chart Analysis for Categorical Column
sns.set(style="whitegrid")
print('Analyze the Distribution :\n')
cat_count = df[Categorical_columns].value_counts()
print(f'{cat_count}\n')
for column in Categorical_columns:
    print(f'INSIGHTS on {column}:')
    #1 most common cat
    Most_common = cat_count.idxmax()
    print(f'Most common Category : {Most_common}')
    #2 least commom cat
    Least_common = cat_count.idxmin()
    print(f'Least common Category : {Least_common}\n')
    #3proportion of each cat
    category_proportion = cat_count/len(df)
    print(f'Proportion of {column} : \n')
    print(f'{category_proportion}\n')
    #4 total sales by cat
    total_sales = df.groupby(column)['Total Sales Value'].sum().sort_values(ascending=False)
    print(f'Total sales by category :')
    print(total_sales)
    #bar chart
    df[column].value_counts().plot(kind='bar')
    plt.title(f'Bar chart for {column}')
    plt.xlabel(column)
    plt.ylabel('count')
    plt.show()
    
```

```python
#Standardization of Numerical Variables
for column in Numerical_columns:
    mu = df[column].mean()
    sigma = df[column].std()
    df[column+'_standardized']=(df[column] - mu)/sigma
    print(df.head())
    
```

```python
#comparization 
plt.figure(figsize=(12,len(Numerical_columns)*9))
for i,column in enumerate(Numerical_columns):
    plt.subplot(len(Numerical_columns),2,2*i + 1)
    sns.histplot(df[column], kde = True,color='blue')
    plt.title(f'Orignal Distribution : {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    # for standardized data
    plt.subplot(len(Numerical_columns),2,2*i + 2)
    sns.histplot(df[column + '_standardized'], kde = True,color='red')
    plt.title(f'Standardized Distribution : {column + "_standardized"}')
    plt.xlabel(column + '_standardized')
    plt.ylabel('Frequency')
```
