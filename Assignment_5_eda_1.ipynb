```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

df = pd.read_csv("D:/Assignments/EDA1/EDA1/Cardiotocographic.csv")
df.head()
```

```python
#task 1
#Now to check if there are any missing values
missing_values = df.isnull().sum()
print("Missing values in each column : ")
print(missing_values)
```

```python
#Haldling the missing values or by using the imputaion meathod
df = df.dropna()
missing_values_2 = df.isnull().sum()
print("Missing values in each column : ")
print(missing_values_2)
```

```python
#To ensure that all the values in columns are numeric if not then convert them 
df = df.apply(pd.to_numeric,errors='coerce')
df.head()
```

```python
#Now to identify if there are any outliers,threats
z_score = df.apply(zscore)
df_cleaned = df[(z_score<3).all(axis=1)]
df = df_cleaned 
df.head()
```

```python
#task 2
#statistical summary
summary = df.describe()
print("Statistical summary of the data set :")
print(summary)
```

```python
#simply
df.describe()
```

```python
#task 3
#data visualization
#first histogram for numerical variables
for column in df:
    skewness = df[column].skew()
    if skewness > 0:
        print(f"The distribution is Right-skwed for {column}.")
    elif skewness < 0:
        print(f"The distribution is Left-skwed {column}.")
    else:
        print(f"The distribution is Approximately symmetric {column}.")
    print(f'Skewness for {column}: {skewness}')
    plt.hist(df[column],bins=30,edgecolor='k')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#for box plot
for column in df:
    plt.figure(figsize=(10,10))
    sns.boxplot(df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#Correlation heat map
print("Correlation Heat map :")
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
#pair plot
print("Pair plot :")
sns.pairplot(df)
plt.show()
```

```python
#pattern recognization
#cheking correlations
correlations = df.corr()
print("High correlation b/w variables :")
high_corr = correlations[(correlations > 0.5)&(correlations != 1)].dropna(how='all').dropna(axis=1,how='all')
print(high_corr)
```
